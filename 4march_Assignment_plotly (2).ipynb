```python
# Answer1
logging.info(" Loading the titanic dataset using the load_dataset function of seaborn.")
logging.info("importing the seaborn and plotly ")

import seaborn as sns
#plotly.express
import plotly.express as epx
titanic_data = sns.load_dataset("titanic")

scatter = epx.scatter(x = titanic_data.age ,y = titanic_data.fare )
scatter.show()


```

```python
# Answer2
logging.info("Using the tips dataset to plot the box plot using the plotly express")

import plotly.express as px 
#loading the tip data 
tip_data = px.data.tips()

#box plot using the plotly.express 

box = px.box(tip_data,x= 'day' ,y= 'total_bill')
box.show()


```

```python
# Answer3
logging.info("Histogram of the tip dataset ")

import plotly.express as px

#loading the tips dataset 
tip_data = px.data.tips()
#plotting the hostogram with px.histogram() method
hist  = px.histogram(tip_data,x= 'sex',y= 'total_bill', pattern_shape='smoker' ,color = 'day')
hist.show()
```

```python
# ANswer4
#Scatter matrix plot
logging.info("using the iris datset to plot scatter matrix ")

import plotly.express as px

iris_Data = px.data.iris()
#scatter_matrix

scatter_mat = px.scatter_matrix(iris_Data,dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"] ,color = 'species')
scatter_mat.show()
```

```python
# ANswer5
# displot
"""
distplot is a function from the Seaborn library used to visualize the distribution of a univariate set of observations
"""

# Example

import plotly.express as px
import numpy as np

# Generate some random data
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# Create a distplot using Plotly express
dist_plot = px.histogram(data, nbins=30, marginal="rug", opacity=0.7, color_discrete_sequence=["blue"])

# Show the distplot
dist_plot.show()


```
