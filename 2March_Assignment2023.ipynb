```python
# Answer2
# Scatter plot:

"""
We can implement the scatter plots while comparing various data variables to determine the connection between dependent and independent variables
The data gets expressed as a collection of points clustered together meaningfully. 
"""
# For Example 

logging.info("Given Data ")
logging.info("import matplotlib.pyplot library")

import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# "plotting the scatter graph using the scatter function

plt.title("Scatter Graph of the random Data")  #tite
plt.xlabel("The is X data")                    #xlabel
plt.ylabel("This is Y data")                    #ylabel
plt.scatter(x,y)
plt.show()



```

```python
# Answer3
# Four plot using the subplot
import matplotlib.pyplot as plt
import numpy as np

logging.info("define the data for the four lines")
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

logging.info("create a 2x2 subplot")
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# plot the first line in the top left subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# plot the second line in the top right subplot
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# plot the third line in the bottom left subplot
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# plot the fourth line in the bottom right subplot
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

logging.info("add a common x label and y label to the entire figure")
fig.text(0.5, 0.04, 'X values', ha='center', va='center')
fig.text(0.06, 0.5, 'Y values', ha='center', va='center', rotation='vertical')

# add a title to the entire figure
fig.suptitle('Four Line Plots using the subplot function')

# show the plot
plt.show()

```

```python
# Answer4
# bar plot
"""
A bar plot is a type of visualization that represents data using rectangular bars graph.
Bar plots are commonly used to compare and display data across different categories or groups.
"""

# It is used because
"""
Bar plots are commonly used to visualize categorical or qualitative data, where the categories are typically displayed along the x-axis.
In a bar plot, the height (or length) of each bar represents the value of the data being plotted. The bars can be colored or filled with patterns to help distinguish between different categories or groups.
Bar plots can be either vertical (where the bars are displayed vertically) or horizontal (where the bars are displayed horizontally).
"""

# Given  Data in the Question
logging.info("import the matplotlib.pyplot ")

import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
# Answer4 
# a

#printing the bar graph
plt.title("Bar Graph between comapany and profit")
plt.xlabel("comapny")
plt.ylabel("profit")
plt.bar(company,profit)
plt.show()
```

```python
# Answer4
# b
# Horizontal Bar Graph

plt.title("Horizontal Bar Graph between comapany and profit")
plt.xlabel("comapny")
plt.ylabel("profit")
#using the barh()
plt.barh(company,profit)
plt.show()
```

```python
# Answe5
# Box plot
"""
A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset
A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum

"""
# It is used because
"""
A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset that shows the distribution of the data based on five summary statistics: the minimum value, the maximum value, the median, and the first and third quartiles. 
Box plots are used to visually compare the distribution of different datasets or to identify outliers.

"""

# Boxplot of the given data

import matplotlib.pyplot as plt
import numpy as np 
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

logging.info("create the box plot")
#passing the ddat as list
plt.xlabel("Box")
plt.ylabel("Value")
plt.title("Box plot graph")
plt.boxplot([box1,box2])
#printing the box plot
plt.show()
```
