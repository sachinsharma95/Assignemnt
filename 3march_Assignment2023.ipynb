```python
# Answer1
# Seaborn
"""
Seaborn is a library that uses Matplotlib underneath to plot graphs. It will be used to visualize random distributions.
"""
# Five plots that we can plot using the seaborn library with their uses 
"""
 1 Scatter Plot:

Used to visualize the relationship between two continuous variables.
Each point represents an observation in the data.
Useful for detecting outliers and identifying patterns.

2 Line Plot:

Used to visualize the trend of a continuous variable over time or other continuous variable.
Typically used to represent time-series data.
Useful for identifying trends and seasonality in the data.

3 Bar Plot:

Used to visualize the distribution of a categorical variable.
Each bar represents the count or percentage of observations in each category.
Useful for comparing the distribution of a variable across different categories.


4 Box Plot:

Used to visualize the distribution of a continuous variable.
Shows the quartiles of the data as well as any outliers.
Useful for identifying the spread, skewness, and outliers in the data.

5 Heatmap:

Used to visualize the correlation between multiple variables.
The colors in the heatmap represent the strength of the correlation between pairs of variables.
Useful for identifying patterns and relationships between variables in a large dataset.

"""
```

```python
# Ansswer2
# lineplot

import seaborn as sns

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Plot a line plot with "timepoint" on x-axis and "signal" on y-axis for different events and regions
logging.info("using the style and hue for different region")
sns.lineplot(x="timepoint", y="signal", data=fmri_data, style="event", hue="region")

```

```python
# ANswer3

import seaborn as sns

logging.info("load the titanic dataset ")

titanic_data = sns.load_dataset("titanic")
titanic_data
```

```python
# ANswer3
# boxplot 

# Plot a box plot with "pclass" on x-axis and "age" on y-axis
sns.boxplot(x="pclass", y="age", data=titanic_data)


# Plot a box plot with "pclass" on x-axis and "fare" on y-axis
sns.boxplot(x="pclass", y="fare", data=titanic_data)
```

```python
# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
# parameter for the 'cut' column of the diamonds dataset.
```

```python
# ANswer4
# histogram

import seaborn as sns

logging.info("laod the diamonds dataset from the seaborn")

diamonds_data = sns.load_dataset("diamonds")
diamonds_data

#histogram for the price column
#using the hue for the cut column
sns.displot(x ='price',hue = 'cut',kind = "hist",data =diamonds_data, multiple = 'stack')
```

```python
# ANswer6
import seaborn as sns
logging.info("laod the flights dataset from the seaborn")

flights_data = sns.load_dataset("flights")

# let's use the pivot_table() function to create a pivot table
flights_data = flights_data.pivot('month', 'year', 'passengers')

# plot the heat map
sns.heatmap(flights_data,cmap = "cool)


```
