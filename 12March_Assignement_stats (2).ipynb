```python

# answer6
logging.info("Chi-square distribution with 10 degrees of freedom")

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# degrees of freedom
df = 10

# create x-axis values
x = np.linspace(0, 30, 100)

# create chi-square distribution with df degrees of freedom
dist = chi2(df)   


# plot the chi-square distribution
plt.plot(x, dist.pdf(x),'r')

# shade the area corresponding to a chi-square statistic of 15

x_shade = np.linspace(15, 30, 100)
y_shade = dist.pdf(x_shade)
plt.fill_between(x_shade, y_shade, color='grey')

# label the axes
plt.xlabel('Chi-square statistic')
plt.ylabel('Probability density')
plt.title('Chi-square distribution with 10 degrees of freedom')

# show the plot
plt.show()
```

## Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.

### Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.
### Solution given_below in python code

```python
# Answer11
# Testing the population mean
"""
1 Null hypothesis: 
the population mean is not significantly different from 70.

"""
# 2 Alternate hypothesis(Ha)
"""
the population mean is significantly different from 70.

"""

# testing and crtical_value comparison

import scipy.stats as stat
import numpy as np

# data given in the question 
sample_size = 30
sample_mean = 72
sample_dev =10
population_mean = 70
alpha =0.05


# t-test statics
t_test = (sample_mean - population_mean) / (sample_dev / np.sqrt(sample_size))
print("The T-test is :",t_test)
# calculating the Crtical_value
critical_val = stat.t.ppf(alpha/2,df = sample_size-1)
print("the critical-value is ",critical_val, abs(critical_val))


# hypo test 
if(t_test<critical_val and t_test>(critical_val)):
    print("Reject the Null hypothesis: the population mean is significantly different from 70")
else:
    print("We fail to Reject the Null : the population mean is not significantly different from 70.")
```
