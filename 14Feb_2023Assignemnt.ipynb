```python
# Ans3

# 1 run() : Method representing the thread’s activity.
"""
This method is responsible for executing the target method.
It is called when the thread is started using the start() method. 
In other words, this method contains the code that will run in the new thread.
"""
# For example:
import threading   #importing the module threading

def message():
    print( " We are using the run function of multithreading")
    
# Thread constructor crete anew Thread
t = threading.Thread( target = message)
# start the thread 
t.run()

# 2 start():
"""
Start the thread’s activity.
It must be called at most once per thread object 
It arranges for the object’s run() method to be invoked in a separate thread of control. 
"""
# for example:
def message1():
    print( " We are using the start method..........")
    
# Thread constructor crete anew Thread
t = threading.Thread( target = message1)
# run the thread 
t.start()   # start method

# 3  join(timeout=None)

"""
Wait until the thread terminates.
This method is used to wait for the thread to complete its execution. When this method is called, the calling thread will wait for the target thread to finish before continuing its own execution.. 
This blocks the calling thread until the thread whose join() method is called terminates – either normally or through an unhandled exception – or until the optional timeout occurs.
"""
# Example of the join:

def message3():
    print( " We are using the join method..........")
    
# Thread constructor crete anew Thread
t = threading.Thread( target = message3) 
t.start()   # start method
t.join()    # wait for the thread to complete



# 4 isAlive():
"""
 This method is used to check whether the thread is currently executing or not. 
 It returns True if the thread is currently running, and False otherwise.
 """
# Example of the isAlive():

def message4():
    print( " We are using the isalive method..........")
    
# Thread constructor crete anew Thread
t = threading.Thread( target = message4) 
print(t.is_alive())  # return False because thread not started
t.start()   # start method
print(t.is_alive())  
t.join()    # wait for the thread to complete

```

```python
import threading

logging.info("Function to print a list of squares as per the question")
def print_squares():
    squares = [x**2 for x in range(1, 10)]
    print("Squares:", squares)

logging.info("Function to print a list of cubes")
def print_cubes():
    cubes = [x**3 for x in range(1, 10)]
    print("Cubes:", cubes)

logging.info("Create the threads")
t1 = threading.Thread(target=print_squares)
t2 = threading.Thread(target=print_cubes)

logging.info("Start the threads")
t1.start()
t2.start()

# Wait for the threads to finish
t1.join()
t2.join()

#message to indiacte that program has completed 
print("Program Completed ...")
```
