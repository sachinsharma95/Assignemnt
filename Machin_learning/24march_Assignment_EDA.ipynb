## Handling missing values

- ### Mean Value Imputation :
    this is generally used when dataset is normally distrubuted
    

- ### Median Value Imputaion:
    It is generally used when the you have outliers in the dataset.
    
- ### Mode value Imputations:
    It is Generally used when the dataset have the categorical datset
    
    
-----

## Advnateges and Disadvantages of the All techniques

## 1 Mean Value Imputation 

### Advantages:

- Simple and easy to implement.
- Preserves the mean of the data, which is important in some applications.
- Can work well if the missing data is missing at random (MAR).


### Disadvantages:

- Can introduce bias into the data if the missing data is not missing at random (NMAR).
- Can result in unrealistic or invalid values if the distribution of the data is not normal.

## 2 Median Value Imputation:
### Advantages:

- Simple and easy to implement.
- Robust to outliers.
- Can work well if the missing data is NMAR.

### Disadvantages:

- Can result in a loss of information if the distribution of the data is not symmetrical.
- May not be suitable for continuous data with a wide range of values.

## 3 Mode Value Imputation:
### Advantages:

- Simple and easy to implement.
- Suitable for categorical data.
- Can work well if the missing data is MAR.

### Disadvantages:

- May not be suitable for continuous data. it is used for categoricsl datset

- Can result in a loss of information if there are multiple modes.

## To improve normality for non-normally distributed features, there are several transformations that can be applied

- ### Logarithmic transformation: 
    This is commonly used for right-skewed data, where large values occur less frequently. Taking the logarithm of these values can help to spread out the data and make it more normally distributed.
    
- ### Square root transformation: 
    This can be used for data that are positively skewed but not as severely as with a logarithmic transformation
    
- ### Box-Cox transformation: 
   This is a more general transformation that can be used for both positively and negatively skewed data. It involves raising the data to a power (lambda) that is chosen to make the transformed data more normally distributed.
   
- ### Reciprocal transformation: 
   This can be used for data that are heavily skewed towards small values. Taking the reciprocal of the data can help to spread out the values and make the distribution more normal.
   
- ### Exponential transformation: 
    This can be used for data that are negatively skewed, where small values occur less frequently. Taking the exponential of these values can help to spread out the data and make it more normally distributed.

### Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?

 # Answer 6 
 
 
 
 
 
 

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# load dataset
df = pd.read_csv('winequality-red.csv')

# separate target variable from features
X = df.drop('quality', axis=1)
y = df['quality']


# scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)



# perform PCA
pca = PCA(n_components=8)

X_pca = pca.fit_transform(X_scaled)

# calculate explained variance ratio
explained_variance = pca.explained_variance_ratio_

# print cumulative explained variance ratio for each number of components
cumulative_variance = np.cumsum(explained_variance)
print(cumulative_variance)
```

```python
# plot scree plot
plt.plot(range(1,9), cumulative_variance, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.show()
```

This will show a graph with the cumulative explained variance ratio for each number of components. The elbow point on the graph indicates the minimum number of components required to explain a significant portion of the variance in the data. In this case, we can see that the elbow point is at around 6 components, confirming our previous analysis.
